```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = "https://www.tickertape.in/etfs/icici-prudential-nifty-100-low-vol-30-etf-IICI"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
csp_price = soup.find("title")
csp_price = csp_price.text
csp_price = re.findall(r'\₹\s+(\d+\.\d+)', csp_price)[0]
print(csp_price)
```

```python
Main = pd.read_csv("all2.csv")

# remove unwanted Columns
Copy_1 = Main.copy()

# create new Columns (Today's Current Price)
csp_list = []
for url in Copy_1['links']:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    csp_price = soup.find("title")
    csp_price = csp_price.text
    print(csp_price)
    try:
        csp_price = re.findall(r'\₹\s+(\d+\.\d+)', csp_price)[0]
    except:
        try:
            csp_price = re.findall(r'₹[^\d]*(\d{1,3}(?:,\d{3})*(?:\.\d{2})?)', csp_price)[0]
            csp_price = csp_price.replace(",","")
        except:
            pass
    
    # # change float data into int Data
    csp_price = int(float(csp_price))
    csp_list.append(csp_price)
print("out_loop")

```

```python
try:
    Copy_1 = Copy_1.drop("scraper_links",axis=1)
except:
    pass

Copy_1.loc[:,'csp'] = csp_list
Copy_1

# # change float data into int Data

Copy_1['Low'] = Copy_1['Low'].astype(int)
Copy_1['High'] = Copy_1['High'].astype(int)
print(Copy_1)
```

```python
#####  
avg_df = pd.read_csv("yes2.csv")
print(avg_df)
for index,rec in Copy_1.iterrows():
    average = np.linspace(rec["High"],rec["Low"],9)
    avg_df.loc[index] = [rec["Names"],rec["High"],rec["Low"],rec["csp"],average[0],average[1],average[2],average[3],average[4],average[5],average[6],average[7],average[8]] 
#     avg_df = avg_df.append(new_row, ignore_index=True)

# Print the DataFrame
# print(avg_df)
# print(average)
# # change float data into int Data

avg_df['AVG_100'] = avg_df['AVG_100'].astype(int)
avg_df['AVG_87'] = avg_df['AVG_87'].astype(int)
avg_df['AVG_75'] = avg_df['AVG_75'].astype(int)
avg_df['AVG_62'] = avg_df['AVG_62'].astype(int)
avg_df['AVG_50'] = avg_df['AVG_50'].astype(int)
avg_df['AVG_37'] = avg_df['AVG_37'].astype(int)
avg_df['AVG_25'] = avg_df['AVG_25'].astype(int)
avg_df['AVG_12'] = avg_df['AVG_12'].astype(int)
avg_df['AVG_0'] = avg_df['AVG_0'].astype(int)
avg_df['csp'] = avg_df['csp'].astype(int)

pd.set_option('display.max_rows', None)
print(avg_df)
# avg_df.to_csv('data011.csv', index=False)
```

```python
# Analysis 
result = []
for index,record in avg_df.iterrows():
#     print(record["csp"])
    if record["csp"]<record["AVG_0"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":9,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_12"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":8,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_25"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":7,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_37"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":6,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_50"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":5,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_62"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":4,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_75"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":3,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_87"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":2,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
        
    elif record["csp"]<record["AVG_100"]:
        Profit_Percentage = ((record["High"]-record["csp"])/record["csp"])*100
        csp = record["csp"]
        Profit = record["High"]-record["csp"]
        nm = record['Names']
        result.append({"S_No":1,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
    else:
        result.append({"S_No":0,"Name":nm,"csp":csp, "Profit_Percentage":Profit_Percentage, "Profit" : Profit})
       
```
